```python
from tensorflow.keras.models import load_model
from PIL import Image
import numpy as np
import ast
import matplotlib.pyplot as plt
```

```python
model = load_model('saved_models/plant_disease_classification.h5')
```

```python
img = Image.open('Manual_Testing_Images/testimg.jpg')
img = img.resize((224, 224))
plt.imshow(img)# Resize the image while maintaining aspect ratio
img = np.array(img)  # Convert image to NumPy array
img = img.astype('float32')  # Convert pixel values to float32
# img = img[..., ::-1]  # Convert from BGR to RGB (if necessary)
# img /= 255.0  # Normalize pixel values
img = np.expand_dims(img, axis=0) 
```

```python
prediction = model.predict(img)
class_indices = np.argmax(prediction, axis=1)
print(class_indices)
with open('classeslist.txt','r') as f:
    classes_dict=ast.literal_eval(f.read())
predicted_class = classes_dict[class_indices[0]]
```
